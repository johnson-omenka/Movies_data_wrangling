## Introduction
This is a dataset that contains various attributes of movies. The aim is to clean the dataset and perform analyses to draw inferences

## Importing the Packages

```python
# Import all packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data and assigning it to a variable

wrangling = pd.read_csv('wrangling.csv')
```

```python
# To see the file

wrangling
```

```python
# To check the structure of the dataset

wrangling.info()
```

```python
# To delete columns that are not needed


wrangling.drop(['Unnamed: 0', 'Unnamed: 8'], axis = 1, inplace = True)
```

```python
# To clean the column names

wrangling = wrangling.rename(columns={'Original titlÊ': 'Title','Genrë¨': 'Genre', ' Votes ': 'Votes'})
```

```python
# To replace spaces in column names with underscore(_)

wrangling.columns = wrangling.columns.str.replace(' ', '_')
```

```python
# To take only the first genre and strip the rest in the genre column
wrangling.Genre = wrangling.Genre.str.split(',', expand =True)[0]
```

```python
wrangling
```

```python
# To clean the country column and make the country names consistent
# First we will use the value count function to see the countries

wrangling.Country.value_counts()
```

```python
# Next is to write a code to make the country names consistent

wrangling['Country'] = wrangling['Country'].replace({"US.":"USA", "US": "USA", "Italy1": "Italy", "West Germany": "Germany", "New Zesland": "New Zealand", "New Zeland": "New Zealand" })
```

```python
wrangling.info()
```

```python
# The release year has some errors in the date which would be cleaned also

wrangling.loc[70, 'Release_year'] ='10 August 1950'
wrangling.loc[83, 'Release_year'] ='9 December 1983'
```

```python
# To check the location of a particular error

wrangling[wrangling["Release_year"] == "1976-13-24"]
```

```python
# Using the loc function to reassign the correct date

wrangling.loc[84, 'Release_year'] ='8 February 1976'
```

```python
# Code to convert datatype of release year from object to datetime

wrangling["Release_year"] = wrangling["Release_year"].astype('datetime64[ns]')
```

```python
wrangling
```

```python
# Now to clean the Score column
# We would first check the value count and know which number to clean

wrangling.Score.value_counts()
```

```python
# Code to replace the score values and reassign to the Score column

wrangling['Score'] = wrangling['Score'].replace({"8,7e-0":"8.7", "8.7.": "8.7", "8:08": "8.0", "8..8": "8.8"})
```

```python
# another method

wrangling.replace("8,9f", "8.9", inplace=True)
wrangling.replace("9,.0", "9.0", inplace=True)
wrangling.replace("8,6", "8.6", inplace=True)
wrangling.replace("9", "9.0", inplace=True)
wrangling.replace("8", "8.0", inplace=True)

# this method was used because the punctuation mark(,) caused error in the dictionary in cell 29
```

```python
# now we change the datatype of Score from int to float

wrangling['Score'] = wrangling['Score'].astype(float)
```

```python
# next is to change the datatype of Income from object to int but we have to get rid of the dollar sign

wrangling.Income = wrangling.Income.str.split(' ',expand= True)[1]
```

```python
wrangling
```

```python
# we want to clean the figure in position 3

wrangling.loc[3, 'Income'] ='408035783'
```

```python
# to delete an empty row causing an error

wrangling = wrangling.drop(13)
```

```python
# to change the data type of Income from object to int

wrangling['Income'] = wrangling['Income'].astype(str).astype(int)
```

```python
# to clean the Votes column, and then change it to float 

wrangling['Votes'] = wrangling['Votes'].str.replace('.','')
wrangling['Votes'] = wrangling['Votes'].astype(str).astype(int)
```

```python
# next is to clean the Duration column and change the dataype from object to int

wrangling.loc[4, 'Duration'] ='154'
wrangling.loc[6, 'Duration'] ='195'
wrangling.loc[9, 'Duration'] ='139'
wrangling.loc[11, 'Duration'] ='178'
wrangling.loc[15, 'Duration'] ='179'
wrangling.loc[16, 'Duration'] ='124'
wrangling.loc[17, 'Duration'] ='146'
wrangling.loc[18, 'Duration'] ='133'
```

```python
# change the dataype from object to float

wrangling['Duration'] = wrangling['Duration'].astype(str).astype(float)
```

```python
wrangling
```

## Now with this cleaning, we can plot charts to make inference

```python
# To find out the genre that is most common

plt.figure(figsize = (12, 6))
plt.xticks(rotation = 45)
plt.xlabel('Genre', size = 12)
plt.ylabel('Count')
plt.title('Genre Distribution', size = 14, fontweight='bold')

ax = sns.countplot(x ='Genre', data = wrangling,color = 'black')
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light')


plt.show()
```

```python
# To find out the country that produces more movies

plt.figure(figsize = (12, 6))
plt.xticks(rotation = 45)
plt.xlabel('Country', size = 12)
plt.ylabel('Count')
plt.title('Country Distribution', size = 14, fontweight='bold')

ax = sns.countplot(x ='Country', data = wrangling,color = 'black')
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light')


plt.show()
```
